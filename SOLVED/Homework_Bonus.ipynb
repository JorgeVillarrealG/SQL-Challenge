```python
import pandas as pd 
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
from pwsql import pwsql
```

```python
DATABASE_URI = f'postgres+psycopg2://postgres:{pwsql}@localhost:5432/EmployeeSQL'
```

```python
engine = create_engine(DATABASE_URI)
```

```python
data= pd.read_sql("SELECT * FROM salaries",engine)
```

```python
#Salary Histogram
plt.hist(data["salary"])
plt.xlabel("Salary$")
plt.ylabel("Count")
plt.show()
```

```python
#Salary by Title Bar Chart
salary= pd.read_sql("SELECT a.emp_no, b.title, a.salary FROM salaries AS a INNER JOIN titles AS b ON a.emp_no = b.emp_no", engine)
```

```python
salary_df=pd.DataFrame(salary.groupby("title")["salary"].mean())
```

```python
salary_df.plot(kind="bar")
plt.title("Salary Averegar by Title")
plt.xlabel("Title")
plt.ylabel("Salary")
plt.show()
```
